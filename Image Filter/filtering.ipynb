```
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

```
img = cv2.imread('image.jpg', 0)
```

```
def create_noise(img, kind):
  h,w = img.shape
  noise = np.random.normal(0, 40, (h,w))
  return img + noise
```

```
gause_noise_img = create_noise(img, 'gause')
```

```
gause_noise_img = np.int16((gause_noise_img - np.min(gause_noise_img)) / (np.max(gause_noise_img) - np.min(gause_noise_img)) * 255)
```

```
fig, ax = plt.subplots(1,2)

fig.set_size_inches(15,15)

ax[0].imshow(img)
ax[0].set_title('origin image')

ax[1].imshow(gause_noise_img)
ax[1].set_title('gause nois image')

plt.show()
```

```
mean_filter = np.ones((3,3)) / 9
```

```
restore_img = convolve2d(gause_noise_img, mean_filter)
```

```
fig, ax = plt.subplots(1,3)

fig.set_size_inches(20,20)

ax[0].imshow(img)
ax[0].set_title('origin image')

ax[1].imshow(gause_noise_img)
ax[1].set_title('gause nois image')

ax[2].imshow(restore_img)
ax[2].set_title('restore image')

plt.show()
```
